```python
# Prediction Function
def predict_image(model, image_path):
    """
    Predicts if there is a tumor in the MRI image.

    Args:
        model: Trained model.
        image_path (str): Path to the MRI image.

    Returns:
        str: Prediction result - 'Tumor Detected' or 'No Tumor Detected'.
    """
    img = cv2.imread(image_path)
    if img is None:
        print("Error loading image.")
        return None
    # Preprocess the image
    img_resized = cv2.resize(img, IMG_SHAPE)
    img_rgb = cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB)
    img_normalized = img_rgb / 255.0
    img_input = np.expand_dims(img_normalized, axis=0)

    # Make prediction
    prediction = model.predict(img_input)
    print("Raw prediction value:", prediction[0][0])
    result = "Tumor Detected" if prediction[0][0] >= 0.5 else "No Tumor Detected"

    # Display the image and result
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.title(result)
    plt.show()

    return result


# Example Usage of the Prediction Function

# Replace the image paths with your own test images
image_path = "/content/R.jpg"
prediction1 = predict_image(best_vgg_model, image_path)
print(f"Prediction using VGG16 model: {prediction1}")

image_path = "/content/nei.jpg"
prediction2 = predict_image(best_vgg_model, image_path)
print(f"Prediction using VGG16 model: {prediction2}")
```

```python
# Install kaggle
!pip install -q kaggle

# Upload kaggle.json
from google.colab import files
files.upload()

# Set up kaggle API
!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json

# Download and unzip dataset
!kaggle datasets download ayushdutta2512/brain-tumor-datasetfor-binary-classification
!unzip brain-tumor-datasetfor-binary-classification.zip

# Imports and setup
import warnings
warnings.filterwarnings('ignore')

import os
import random
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf

from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPool2D, Dropout
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import VGG16
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score

os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

SEED = 40
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

# Paths and constants
MAIN_DIR = "/content/Brain Tumor Dataset(for binary classification)"
IMG_SHAPE = (128, 128)
BATCH_SIZE = 32

# Check dataset
print("Dataset folders:", os.listdir(MAIN_DIR))
for dirpath, dirnames, filenames in os.walk(MAIN_DIR):
    print(f"{len(dirnames)} directories and {len(filenames)} images in {dirpath}")

# Random image viewer
def view_random_image():
    subdirs = ['yes', 'no']
    subdir = random.choice(subdirs)
    target_folder = os.path.join(MAIN_DIR, subdir)
    random_image = random.choice(os.listdir(target_folder))
    img_path = os.path.join(target_folder, random_image)
    img = cv2.imread(img_path)
    if img is None:
        print(f"Error loading image: {img_path}")
        return
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.axis('off')
    plt.title(f"Image shape: {img.shape}")
    plt.show()

view_random_image()

# Data preparation
datagen = ImageDataGenerator(rescale=1/255., validation_split=0.5)

train_data = datagen.flow_from_directory(
    MAIN_DIR,
    target_size=IMG_SHAPE,
    batch_size=BATCH_SIZE,
    class_mode="binary",
    shuffle=True,
    subset="training",
    seed=SEED
)

test_data = datagen.flow_from_directory(
    MAIN_DIR,
    target_size=IMG_SHAPE,
    batch_size=BATCH_SIZE,
    class_mode="binary",
    shuffle=False,
    subset="validation",
    seed=SEED
)

# CNN model
cnn_model = Sequential([
    Conv2D(64, 3, activation='relu', input_shape=(IMG_SHAPE[0], IMG_SHAPE[1], 3)),
    Conv2D(32, 3, activation='relu'),
    MaxPool2D(2),
    Conv2D(32, 3, activation='relu'),
    Conv2D(16, 3, activation='relu'),
    MaxPool2D(2, padding='same'),
    Flatten(),
    Dense(1, activation='sigmoid')
])

cnn_model.compile(loss=BinaryCrossentropy(), optimizer=Adam(), metrics=["accuracy"])

# Callbacks
cnn_checkpoint = ModelCheckpoint('best_model1.h5', monitor='val_accuracy', save_best_only=True, mode='max')
early_stop = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

steps_per_epoch = train_data.samples // BATCH_SIZE
validation_steps = test_data.samples // BATCH_SIZE

# Train CNN
history_cnn = cnn_model.fit(
    train_data,
    epochs=10,
    steps_per_epoch=steps_per_epoch,
    validation_data=test_data,
    validation_steps=validation_steps,
    callbacks=[cnn_checkpoint, early_stop]
)

# Plotting function
def plot_curves(history, title_prefix=""):
    loss = history.history["loss"]
    val_loss = history.history["val_loss"]
    accuracy = history.history["accuracy"]
    val_accuracy = history.history["val_accuracy"]
    epochs = range(len(loss))

    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, loss, label="Training Loss")
    plt.plot(epochs, val_loss, label="Validation Loss")
    plt.title(f"{title_prefix} Loss Curves")
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, accuracy, label="Training Accuracy")
    plt.plot(epochs, val_accuracy, label="Validation Accuracy")
    plt.title(f"{title_prefix} Accuracy Curves")
    plt.legend()
    plt.show()

plot_curves(history_cnn, "CNN")

# Evaluate CNN
best_cnn_model = load_model('best_model1.h5')
cnn_preds = best_cnn_model.predict(test_data)
cnn_preds_classes = (cnn_preds > 0.5).astype("int32").flatten()
true_labels = test_data.classes

# CNN Metrics
cnn_accuracy = np.mean(cnn_preds_classes == true_labels)
cnn_precision = precision_score(true_labels, cnn_preds_classes)
cnn_recall = recall_score(true_labels, cnn_preds_classes)
cnn_f1 = f1_score(true_labels, cnn_preds_classes)

# Confusion Matrix for CNN
plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(true_labels, cnn_preds_classes), annot=True, fmt='d', cmap="Blues")
plt.title("CNN Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# VGG16 Model
base_model = VGG16(weights='imagenet', include_top=False, input_shape=(IMG_SHAPE[0], IMG_SHAPE[1], 3))
for layer in base_model.layers:
    layer.trainable = False

vgg_model = Sequential([
    base_model,
    Flatten(),
    Dense(256, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

vgg_model.compile(loss=BinaryCrossentropy(), optimizer=Adam(), metrics=["accuracy"])

vgg_checkpoint = ModelCheckpoint('best_model2.h5', monitor='val_accuracy', save_best_only=True, mode='max')
early_stop_vgg = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Train VGG
history_vgg = vgg_model.fit(
    train_data,
    epochs=10,
    steps_per_epoch=steps_per_epoch,
    validation_data=test_data,
    validation_steps=validation_steps,
    callbacks=[vgg_checkpoint, early_stop_vgg]
)

plot_curves(history_vgg, "VGG16")

# Evaluate VGG
best_vgg_model = load_model('best_model2.h5')
vgg_preds = best_vgg_model.predict(test_data)
vgg_preds_classes = (vgg_preds > 0.5).astype("int32").flatten()

# VGG Metrics
vgg_accuracy = np.mean(vgg_preds_classes == true_labels)
vgg_precision = precision_score(true_labels, vgg_preds_classes)
vgg_recall = recall_score(true_labels, vgg_preds_classes)
vgg_f1 = f1_score(true_labels, vgg_preds_classes)

# Confusion Matrix for VGG
plt.figure(figsize=(5,4))
sns.heatmap(confusion_matrix(true_labels, vgg_preds_classes), annot=True, fmt='d', cmap="Greens")
plt.title("VGG16 Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# Comparison Table
comparison_df = pd.DataFrame({
    "Model": ["Custom CNN", "VGG16"],
    "Accuracy": [cnn_accuracy, vgg_accuracy],
    "Precision": [cnn_precision, vgg_precision],
    "Recall": [cnn_recall, vgg_recall],
    "F1 Score": [cnn_f1, vgg_f1]
})

print("\nModel Performance Comparison:")
print(comparison_df)

# Plot Comparison
comparison_df.set_index("Model").plot(kind='bar', figsize=(10,6), colormap='Set2')
plt.title("Model Comparison Metrics")
plt.ylabel("Score")
plt.ylim(0, 1.05)
plt.grid(True, axis='y')
plt.xticks(rotation=0)
plt.show()
```

```python
# Accuracy and Loss Comparison Table
accuracy_loss_df = pd.DataFrame({
    "Model": ["Custom CNN", "VGG16"],
    "Accuracy": [cnn_accuracy, vgg_accuracy],
    "Loss": [history_cnn.history['val_loss'][-1], history_vgg.history['val_loss'][-1]]
})

print("\nAccuracy and Loss Comparison:")
print(accuracy_loss_df)

# Plotting
accuracy_loss_df.set_index("Model")[["Accuracy", "Loss"]].plot(kind='bar', figsize=(8,5), colormap='coolwarm')
plt.title("Model Accuracy vs Loss")
plt.ylabel("Value")
plt.xticks(rotation=0)
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```
